```python
import pandas as pd
import numpy as np

listings = pd.read_csv(r"D:\a_PERSONAL\airbnb_data_engineering\data\raw\listings.csv")
calendar = pd.read_csv(r"D:\a_PERSONAL\airbnb_data_engineering\data\raw\calendar.csv")
reviews = pd.read_csv(r"D:\a_PERSONAL\airbnb_data_engineering\data\raw\reviews.csv")
```

```python
listings_clean = listings.copy()
reviews_clean = reviews.copy()
calendar_clean = calendar.copy()
```

```python
listings_clean.columns
```

```python
listings.columns.tolist()
```

```python
reviews.columns.tolist()
```

```python
calendar.columns.tolist()
```

```python
date_columns = [
    "host_since",
    "price_quote_checkin_date",
    "price_quote_checkout_date",
    "first_review",
    "last_review",
    "last_scraped",
    "calendar_last_scraped"
]

for col in date_columns:
    listings_clean[col] = pd.to_datetime(
        listings_clean[col],
        errors="coerce"
    )
```

```python
reviews_clean["date"] = pd.to_datetime(
    reviews_clean["date"],
    errors="coerce"
)
```

```python
calendar_clean["date"] = pd.to_datetime(
    calendar_clean["date"],
    errors="coerce"
)
```

```python
listings_clean["price"].head()
```

```python
listings_clean["price"] = (
    listings_clean["price"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

listings_clean["price"] = pd.to_numeric(
    listings_clean["price"],
    errors="coerce"
)
```

```python
listings_clean["price_quote_total_price"].head()
```

```python
listings_clean["price_quote_total_price"] = (
    listings_clean["price_quote_total_price"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

listings_clean["price_quote_total_price"] = pd.to_numeric(
    listings_clean["price_quote_total_price"],
    errors="coerce"
)
```

```python
listings_clean["price_quote_price_per_night"].head()
```

```python
listings_clean["price_quote_price_per_night"] = (
    listings_clean["price_quote_price_per_night"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

listings_clean["price_quote_price_per_night"] = pd.to_numeric(
    listings_clean["price_quote_price_per_night"],
    errors="coerce"
)
```

```python
listings_clean["estimated_revenue_l365d"].head()
```

```python
listings_clean["estimated_revenue_l365d"] = (
    listings_clean["estimated_revenue_l365d"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

listings_clean["estimated_revenue_l365d"] = pd.to_numeric(
    listings_clean["estimated_revenue_l365d"],
    errors="coerce"
)
```
